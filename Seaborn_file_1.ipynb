#### Histogram:
##### Histograms represent the data distribution through bins    
distplot(df['numerical_column'],kde=, hist=)

```python
sns.distplot(df['petal_length'])
plt.show()
```

```python
sns.distplot(df['petal_length'],kde=False)
plt.show()
```

```python
sns.distplot(df['petal_length'],hist=False)
plt.show()
```

#### Scatter plot: (Join_plot)
##### It show the relationship between the two variable's
joinplot(x ='numercal_cloumn_name',y ='numercal_cloumn_name',data=)

```python
sns.jointplot(x ='sepal_length',y = 'petal_length',data=df)
plt.show()
```

```python
sns.jointplot(x ='sepal_length',y = 'petal_length',data=df,hue='species')
plt.show()
```

#### Multiple plot:
##### It show the realtionship of all columns with each other
pairplot(data,hue ='categorical_column,diag_kind=, kind=, palette=)
